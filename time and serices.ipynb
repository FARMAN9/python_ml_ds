```python
df=yf.download('AAPL',start='2017-01-01',end='2023-12-30')
df
print(df.loc['2023-01-02'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.title('apple stocks price 2017 to 2023')
plt.xlabel('year')
plt.ylabel('price')
plt.plot(df['Adj Close'])


```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.title('Apple Stocks High and Low Prices (2017-2023)')
plt.xlabel('Year')
plt.ylabel('Price')
plt.plot(df['High'], label='High Price')
plt.plot(df['Low'], label='Low Price')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.title('apple stocks close price 2017 to 2023')
plt.xlabel('year')
plt.ylabel('price')
plt.plot(df['Close'])
```

```python
plt.figure(figsize=(15,5))
plt.title('Apple Stocks  Prices in 2000 and 2023 ')
plt.xlabel('Year')
plt.ylabel('Price')


plt.plot(df.loc['2000']['Adj Close'], label='High Price')
plt.plot(df.loc['2023']['Adj Close'], label='Low Price')

plt.legend()
plt.show()
```
