```python
#groupby
data={"gender":['male','female','female','male','female'],'score':[1,3,2,4,2]}
data 
```

```python
df=pd.DataFrame.from_dict(data)
df
```

```python
df.groupby(df['gender']).mean()
```

```python
df.groupby(df['gender']).max()
```

```python
df.groupby(df['gender']).min()
```

```python
df.groupby(df['gender']).sum()
```

```python
df.groupby(df['gender']).median()
```

```python
df.groupby(df['gender']).count()
```
